# Introduction to the Tutorial Series on XAI in Chemistry

Welcome to this tutorial series on *Explainable Artificial Intelligence (XAI)* applied to molecular property prediction. These tutorials serve as a practical companion to our review paper, providing hands-on examples of various explanation methods discussed in the literature.

The series is designed for **chemistry practitioners** who are interested in understanding how machine learning models make predictions about molecular properties. Whether you are working in drug discovery, materials science, or computational chemistry, these tutorials will equip you with the practical skills to generate and interpret explanations for your own models.

Each tutorial focuses on a specific category of explanation methods, progressing from simpler feature-based approaches to more sophisticated structure-based and counterfactual techniques. By the end of this series, you will have a comprehensive toolkit for making your molecular property prediction models more transparent and interpretable.

## Prerequisites

These tutorials assume **basic familiarity with Python programming**. You should be comfortable with fundamental concepts such as variables, functions, loops, and working with common data structures like lists and dictionaries. Experience with libraries such as NumPy and Pandas is helpful but not strictly required, as we introduce relevant functionality as needed.

This is not a programming course. The focus is on understanding and applying XAI methods rather than teaching Python fundamentals. If you find yourself struggling with the code, we recommend reviewing introductory Python resources before continuing.

Prior exposure to machine learning concepts and basic chemistry or cheminformatics knowledge will be beneficial for understanding the context and significance of the explanations, but we aim to make the tutorials accessible to those new to these areas as well.

## How These Tutorials Are Structured

Each tutorial in this series follows a consistent format to help you navigate the material efficiently. This section explains the conventions and markers used throughout.

### Dataset and Package Information

At the beginning of each tutorial, you will find two standard sections that provide important context:

**📦 Packages.** This section lists the Python packages required for the tutorial, along with links to their documentation. We recommend reviewing these resources if you want to explore the functionality beyond what is covered in the tutorials.

### Section Markers

Throughout the tutorials, different types of content are marked with distinctive emoji indicators to help you identify their purpose at a glance:

- **💡 Concept Sections** &mdash; Introduce theoretical background and motivate why a particular approach is useful. These sections provide the conceptual foundation before diving into implementation details.

- **📚 Method Sections** &mdash; Cover specific explanation techniques and their implementations. These are the core technical content of each tutorial, where you will learn to apply XAI methods to molecular data.

- **🔬 Discussion Sections** &mdash; Disucss the overall content and takeaway messages at the end of the individual tutorial notebooks.

- **📊 Plot Interpretations** &mdash; Explain how to read and interpret the visualizations generated by the code. Understanding these plots is crucial for extracting meaningful insights from the explanations.

- **🔍 Observations** &mdash; Discuss what kinds of observations might be gleaned from certain results and explanations.

- **📝 Notes** &mdash; Highlight important caveats, tips, or additional context that may be helpful. Pay attention to these when troubleshooting or adapting the code for your own use cases.

Section headers follow the format `## **N.M** $\cdot$ 💡 Section Title` where `N.M` indicates the section number within the tutorial.

## A Brief Word on Explainable AI

Explainable AI (XAI) encompasses a range of techniques designed to make machine learning models more transparent and interpretable. In the context of molecular property prediction, XAI methods help us understand *why* a model predicts that a particular compound will have certain properties &mdash; not just *what* the prediction is.

This understanding is valuable for multiple reasons: it helps build trust in model predictions, enables domain experts to validate whether the model has learned chemically meaningful patterns, and can even reveal new insights about structure-property relationships. The tutorials in this series will introduce you to the practical application of these methods, leaving the detailed theoretical foundations to the accompanying review paper.
